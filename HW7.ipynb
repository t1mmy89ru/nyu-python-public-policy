```
#the indicator of interest will be incidence of crime by borocd#
crime = df1[['crime_count','borocd']]
print(crime)
```

```
#We'll filter by complaints related to streetlight#
count_streetlight = df[df.complaint_type == 'Street Light Condition'].groupby('community_board').size().reset_index(name='count_of_311_requests')
print(len(count_streetlight))
print(count_streetlight)
```

```
valid_borocds = ['101','102','103','104','105','106','107','108','109','110','111','112','201','202','203','204','205','206','207','208','209','210','211','212','301','302','303','304','305','306','307','308','309','310','311','312','313','314','315','316','317','318','401','402','403','404','405','406','407','408','409','410','411','412','413','414','501','502','503']

```

```
#create a borocd column#

def recode_borocd_counts(row):
  if row.community_board[0] != 'U':
    if 'MANHATTAN' in row.community_board:
      return '1' + row.community_board[0:2]
    elif 'BRONX' in row.community_board:
      return '2' + row.community_board[0:2]
    elif 'BROOKLYN' in row.community_board:
      return '3' + row.community_board[0:2]
    elif 'QUEENS' in row.community_board:
      return '4' + row.community_board[0:2]
    elif 'STATEN ISLAND' in row.community_board:
      return '5' + row.community_board[0:2]
  else:
    return 'Invalid borocd'

count_streetlight['borocd'] = count_streetlight.apply(recode_borocd_counts, axis=1)

count_streetlight = count_streetlight[count_streetlight.borocd.isin(valid_borocds)]
print(len(count_streetlight))
print(count_streetlight)
```

```
count_streetlight.borocd.astype(int)
crime.borocd.astype(int)
```

```
merged_cd = pd.merge(how='left', left=crime, right=count_streetlight, left_on='borocd', right_on='borocd')
merged_cd
```

```
plt.figure(figsize=(15, 6))

plt.scatter(
    np.array(merged_cd.crime_count), # x: array that fits within the range of the "total" column's values
    np.array(merged_cd.count_of_311_requests), # y
    s=50, # size of marker
    c='#3d03fc', # color hex color picker: https://www.google.com/search?q=color+picker
    edgecolors='black', # outline color of the marker
    alpha=0.5) # transparency. 0 is 100% transparent. 1 is 100% opaque

plt.title('Streetlight 311 Requests vs. Count of Crime')
plt.xlabel('Count of Crime')
plt.ylabel('# of 311 Streetlight Requests')

plt.show()

```

```
import statsmodels.api as sm
y = merged_cd.count_of_311_requests
x = merged_cd.crime_count
# let's add a y intercept to our model
x = sm.add_constant(x)

model = sm.OLS(y, x).fit()
# Print out the statistics
print(model.summary())

```

```
# see model parameters
model.params

# create label for line of best fit
label = 'y = ' + str(round(model.params[0],2)) + ' + ' + str(round(model.params[1],2)) + 'x'
print(label)

```

```
plt.figure(figsize=(15, 6))

plt.scatter(
    np.array(merged_cd.crime_count), # x: array that fits within the range of the "total" column's values
    np.array(merged_cd.count_of_311_requests), # y
    s=50, # size of marker
    c='#3d03fc', # color hex color picker: https://www.google.com/search?q=color+picker
    edgecolors='black', # outline color of the marker
    alpha=0.5) # transparency. 0 is 100% transparent. 1 is 100% opaque

# line of best fit
plt.plot(
    np.array(merged_cd.crime_count), 
    model.params[0] +  model.params[1]*np.array(merged_cd.crime_count),
    c='#02631c', # darker green
    label=label)


plt.title('Streetlight 311 Requests vs. Count of Crime')
plt.xlabel('Count of Crime')
plt.ylabel('# of 311 Streetlight Requests')
plt.legend()

plt.show()

```

My hypothesis was that there would be a positive correlation between 311 requests related to streetlights and the count of crime in a community district. After running the above analysis, however, the association is both miniscule (in fact, negative) and statistically insignificant at the 5% level. Thus, my initial hypothesis is hard to accept.
